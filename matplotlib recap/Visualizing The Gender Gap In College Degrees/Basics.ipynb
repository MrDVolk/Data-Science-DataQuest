```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_light_gray = (171/255, 171/255, 171/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(11, 18))

label_position_dict = {
    1: (8, 87),
    2: (12, 80),
    3: (3, 90),
    16: (90, 10),
    18: (65, 25)
}

def draw_column(column_index, column_names):
    plot_index = column_index
    for name in column_names:
        ax = fig.add_subplot(6, 3, plot_index)
        ax.plot(women_degrees['Year'], women_degrees[name], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[name], c=cb_orange, label='Men', linewidth=3)
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_yticks([0,100])
        ax.axhline(50, c=cb_light_gray, alpha=0.5)
        ax.set_title(name)
        ax.tick_params(bottom="off", top="off", left="off", right="off")
       
        if plot_index in label_position_dict.keys():
            labels_positions = label_position_dict[plot_index]
            ax.text(2005, labels_positions[0], 'Men')
            ax.text(2002, labels_positions[1], 'Women')
        
        if column_names.index(name) != len(column_names)-1:
            ax.tick_params(labelbottom='off')
        
        plot_index += 3

                
draw_column(1, stem_cats)
draw_column(2, lib_arts_cats)
draw_column(3, other_cats)

plt.savefig('gender_degrees.png')
plt.show()
```
