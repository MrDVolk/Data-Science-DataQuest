# Exploring the highest paying majors and the lowest with bar plots

```python
highest_paying_majors = recent_grads.sort_values(by=['Median'], ascending=False).iloc[0:10]
lowest_paying_majors = recent_grads.sort_values(by=['Median'], ascending=True).iloc[0:10]
```
